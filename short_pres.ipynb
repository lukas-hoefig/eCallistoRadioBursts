# plot summed curve 

```python
year = 2017
month = 9
day = 8
time = "05:46:08"
p1 = data.createFromTime(year,month,day,time, observatories.uni_graz, spec_range)
p1.subtract_background()
p1.createSummedCurve(spec_range)

p3 = data.createFromTime(year, month, day, time, observatories.triest, spec_range)
p3.subtract_background()
p3.createSummedCurve(spec_range)

fig, ax = plt.subplots(figsize=(16,9))
p1.plotSummedCurve(ax)
p3.plotSummedCurve(ax)
plt.show()
```

# plot correlation

```python
cor = correlation.Correlation(p1, p3)
cor.calculatePeaks()

fig, ax = plt.subplots(figsize=(16,9))
cor.plotCurve(ax)
print(cor.peaks)
plt.show()
```

# REFERENCE BURST LIST FOR EACH DAY

```python
year = 2017
month = 9
day = 6
def reference(year, month, day):
    file = "../reference/{}_events/{}{}{}events.txt".format(str(year), str(year), str(month).zfill(2), str(day).zfill(2))
    f = open(file)
    lines = f.readlines()
    f.close()
    lines = lines[12:]
    lines
    for line in lines:
        if line=='\n':
            lines.remove(line)

    for line in range(len(lines)):
        lines[line] = lines[line].rsplit(' ')
        lines[line] = list(filter(None, lines[line]))
        try:
            lines[line].remove('+')
        except ValueError:
            pass

    lines = list(filter(lambda line: line[6] == 'RSP', lines))

    print(day,'.', month, '.', year)
    for line in lines:
        print("{:5s}: {:} - {:}".format(line[8].rsplit('/')[0],line[1], line[3]))

reference(year, month, day)
```
